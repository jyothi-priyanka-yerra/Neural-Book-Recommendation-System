```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
books_reviews_df = pd.read_csv('/content/drive/MyDrive/Priyanka/final_book_reviews.csv')
```

```python
ratings_count = books_reviews_df['rating'].value_counts()
plt.figure(figsize=(8, 5))
plt.bar(ratings_count.index, ratings_count.values)
plt.xlabel('Ratings')
plt.ylabel('Number of Ratings')
plt.title('Rating value')
plt.show()
```

```python
user_ratings_count = books_reviews_df['user'].value_counts()
plt.figure(figsize=(8, 5))
plt.hist(user_ratings_count.values, bins=1000)
plt.xlabel('Ratings per user')
plt.ylabel('Number of Users')
plt.title('Distribution of Number of Ratings Given by Users')
plt.show()
```

```python
item_ratings_count = books_reviews_df['item'].value_counts()
plt.figure(figsize=(8, 5))
plt.hist(item_ratings_count.values, bins=1000)
plt.xlabel('Ratings per Item')
plt.ylabel('Number of Items')
plt.title('Number of Ratings per Item')
plt.show()
```
