```python
import json
import pandas as pd
import numpy as np

np.random.seed(42)
```

# User details

```python
books_reviews_df = pd.read_csv('/content/drive/MyDrive/Priyanka/final_book_reviews.csv')
```

```python
books_reviews_df.head()
```

```python
# Get unique user values
user_ids = books_reviews_df['user'].unique()

# Generate random embeddings for each unique user
embedding_size = 300
user_initial_embeddings = {user: list(np.random.rand(embedding_size)) for user in user_ids}
```

```python
len(user_initial_embeddings)
```

# Export embeddings

```python
import pickle
```

```python
with open("/content/drive/MyDrive/Priyanka/user_initial_embeddings.pkl", "wb") as f:
    pickle.dump(user_initial_embeddings, f)
```
