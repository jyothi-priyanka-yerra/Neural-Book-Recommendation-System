```python
import gzip
import json
import pandas as pd
```

### Book Reviews

```python
## Load book reviews
column_names = ["item", "user", "rating", "timestamp"]
books_reviews_df = pd.read_csv('/content/drive/MyDrive/Priyanka/Books.csv', names=column_names)
display(books_reviews_df.head())
print(books_reviews_df.shape)
```

```python
## Remove empty user or item IDs
books_reviews_df = books_reviews_df.dropna(subset=['user', 'item'])
print(books_reviews_df.shape)

## filter out reviews to have recent ones
books_reviews_df.sort_values(by='timestamp', inplace=True)
display(books_reviews_df.head())
print(books_reviews_df.shape)

remove_rows = len(books_reviews_df) // 2
books_reviews_df = books_reviews_df.iloc[remove_rows:]
display(books_reviews_df.head())
print(books_reviews_df.shape)

remove_rows = len(books_reviews_df) // 2
books_reviews_df = books_reviews_df.iloc[remove_rows:]
display(books_reviews_df.head())
print(books_reviews_df.shape)
```

### Books Details

```python
books_metadata = []
with gzip.open("/content/drive/MyDrive/Priyanka/meta_Books.json.gz") as f:
    for l in f:
        books_metadata.append(json.loads(l.strip()))

books_df = pd.DataFrame.from_dict(books_metadata)
display(books_df.head())
print(books_df.shape)
```

```python
## Keep only required columns
books_df = books_df[['category', 'description', 'title', 'brand', 'price', 'asin']]
display(books_df.head())
print(books_df.shape)
```

```python
## Remove ones with empty asin values and remove duplicates
books_df = books_df.dropna(subset=['asin'])
books_df = books_df.drop_duplicates(subset='asin', keep='first')
print(books_df.shape)

## Remove unformatted books
books_df = books_df.dropna(subset=['category', 'title', 'description'], how='all')
books_df = books_df[~books_df.title.str.contains('getTime')] # filter the unformatted rows
print(books_df.shape)
```

```python
books_reviews_df = books_reviews_df[books_reviews_df['item'].isin(books_df['asin'])]
```

```python
prev = len(books_reviews_df)

while len(books_reviews_df) > 0:

  user_review_counts = books_reviews_df['user'].value_counts()
  users_with_min_reviews = user_review_counts[user_review_counts >= 20].index.tolist()
  books_reviews_df = books_reviews_df[(books_reviews_df['user'].isin(users_with_min_reviews))]

  item_review_counts = books_reviews_df['item'].value_counts()
  items_with_min_reviews = item_review_counts[item_review_counts >= 20].index.tolist()
  books_reviews_df = books_reviews_df[(books_reviews_df['item'].isin(items_with_min_reviews))]

  curr = len(books_reviews_df)

  if prev == curr:
    break

  prev = curr
```

```python
print(books_reviews_df.head())
print(books_reviews_df.shape)

print("no. of users: ", books_reviews_df['user'].nunique())
print("no. of books: ", books_reviews_df['item'].nunique())

print("User value counts")
print(books_reviews_df['user'].value_counts())
print("Item value counts")
print(books_reviews_df['item'].value_counts())
```

### Keep common ones from both datasets

```python
books_df = books_df[books_df['asin'].isin(books_reviews_df['item'])]
books_df.reset_index(drop=True, inplace=True)
```

```python
## Final dimensions of both dataframes
print("######### BOOKS METADATA ##############")
display(books_df.head())
print(books_df.shape)

print("######### USER - BOOKS REVIEWS ##############")
display(books_reviews_df.head())
print(books_reviews_df.shape)

print("no. of users: ", books_reviews_df['user'].nunique())
print("no. of books: ", books_reviews_df['item'].nunique())
```

```python
print("User value counts")
print(books_reviews_df['user'].value_counts())
print("Item value counts")
print(books_reviews_df['item'].value_counts())
```
